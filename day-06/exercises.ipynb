# Exercises with WHILE LOOP
Absolutely! Here are a couple of **exercises** to help you practice using `while` loops. These problems are designed to reinforce your understanding and give you hands-on experience. Try solving them on your own, and if you get stuck, feel free to ask for hints or solutions!

### **Exercise 1: Number Guessing Game**
Write a program that generates a random number between 1 and 100 and asks the user to guess it. The program should:
1. Tell the user if their guess is too high or too low.
2. Keep track of the number of guesses.
3. End the game when the user guesses the correct number.

#### Example Output:
```
Guess the number (between 1 and 100): 50
Too high!
Guess the number (between 1 and 100): 25
Too low!
Guess the number (between 1 and 100): 37
Correct! You guessed it in 3 attempts.
```

#### Hints:
- Use the `random` module to generate a random number.
- Use a `while` loop to keep asking for guesses until the user gets it right.
- Use `if` statements to check if the guess is too high or too low.

---

### **Exercise 2: Sum of Numbers**
Write a program that repeatedly asks the user to enter a number. The program should:
1. Keep a running total of the numbers entered.
2. Stop asking for numbers when the user enters `0`.
3. Print the total sum of all the numbers entered.

#### Example Output:
```
Enter a number (0 to quit): 5
Enter a number (0 to quit): 10
Enter a number (0 to quit): 15
Enter a number (0 to quit): 0
The total sum is: 30
```

#### Hints:
- Use a `while` loop to repeatedly ask for input.
- Use a variable to keep track of the running total.
- Exit the loop when the user enters `0`.

---

### **Exercise 3: Password Checker**
Write a program that asks the user to enter a password. The program should:
1. Keep asking for the password until the user enters the correct one.
2. Limit the number of attempts to 3.
3. Print a message if the user exceeds the maximum number of attempts.

#### Example Output:
```
Enter the password: hello
Incorrect password. 2 attempts remaining.
Enter the password: world
Incorrect password. 1 attempt remaining.
Enter the password: python
Access granted!
```

#### Hints:
- Use a `while` loop to keep asking for the password.
- Use a counter to track the number of attempts.
- Use `break` to exit the loop if the correct password is entered.

---

### **Exercise 4: Fibonacci Sequence**
Write a program that generates the Fibonacci sequence up to a specified number of terms. The program should:
1. Ask the user how many terms they want to generate.
2. Print the Fibonacci sequence up to that number of terms.

#### Example Output:
```
How many terms? 7
Fibonacci sequence: 0, 1, 1, 2, 3, 5, 8
```

#### Hints:
- The Fibonacci sequence starts with `0` and `1`. Each subsequent number is the sum of the previous two.
- Use a `while` loop to generate the sequence.
- Use variables to store the current and previous terms.

### **Exercise 5: Multiplication Table**
Write a program that generates the multiplication table for a given number. The program should:
1. Ask the user for a number.
2. Print the multiplication table for that number up to 10.

#### Example Output:
```
Enter a number: 7
7 x 1 = 7
7 x 2 = 14
7 x 3 = 21
...
7 x 10 = 70
```

#### Hints:
- Use a `while` loop to iterate from 1 to 10.
- Use string formatting to print the table neatly.

### **Exercise 6: Factorial Calculator**
Write a program that calculates the factorial of a number. The program should:
1. Ask the user for a number.
2. Calculate and print the factorial of that number.

#### Example Output:
```
Enter a number: 5
Factorial of 5 is: 120
```

#### Hints:
- The factorial of a number `n` is the product of all positive integers less than or equal to `n` (e.g., `5! = 5 x 4 x 3 x 2 x 1 = 120`).
- Use a `while` loop to calculate the factorial.

### **Exercise 7: Countdown Timer**
Write a program that simulates a countdown timer. The program should:
1. Ask the user for a starting number.
2. Count down from that number to 0, printing each number along the way.

#### Example Output:
```
Enter a number: 5
5
4
3
2
1
0
Blast off!
```

#### Hints:
- Use a `while` loop to count down.
- Decrement the number in each iteration.

---

### **Exercise 8: Prime Number Checker**
Write a program that checks if a number is prime. The program should:
1. Ask the user for a number.
2. Print whether the number is prime or not.

#### Example Output:
```
Enter a number: 13
13 is a prime number.
```

#### Hints:
- A prime number is only divisible by 1 and itself.
- Use a `while` loop to check divisibility from 2 up to the square root of the number.

---

### **Exercise 9: Reverse a Number**
Write a program that reverses a number. The program should:
1. Ask the user for a number.
2. Print the reversed number.

#### Example Output:
```
Enter a number: 1234
Reversed number: 4321
```

#### Hints:
- Use a `while` loop to extract digits from the number and build the reversed number.
- Use integer division (`//`) and modulus (`%`) to manipulate the number.

---

### **Exercise 10: Rock, Paper, Scissors Game**
Write a simple Rock, Paper, Scissors game. The program should:
1. Ask the user for their choice (rock, paper, or scissors).
2. Generate a random choice for the computer.
3. Determine the winner and print the result.
4. Allow the user to play again or quit.

#### Example Output:
```
Enter your choice (rock, paper, scissors): rock
Computer chose: scissors
You win!
Play again? (yes/no): no
```

#### Hints:
- Use a `while` loop to keep the game running until the user chooses to quit.
- Use the `random` module to generate the computer's choice.
